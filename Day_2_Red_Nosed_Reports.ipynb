```
def is_safe(levels):
    diffs = [levels[i+1] - levels[i] for i in range(len(levels)-1)]

    # Check increasing or decreasing
    all_increasing = all(d > 0 for d in diffs)
    all_decreasing = all(d < 0 for d in diffs)

    if not (all_increasing or all_decreasing):
        return False

    # Check difference range
    return all(1 <= abs(d) <= 3 for d in diffs)

```

```
safe_count=0
with open("input.txt") as f:
    for line in f:
        line = line.replace("\\", "").strip()
        if not line:
            continue

        try:
            levels = list(map(int, line.split()))
        except ValueError:
            print("Skipping bad line:", line)
            continue

        if is_safe(levels):
            safe_count += 1
    print(safe_count)
```
